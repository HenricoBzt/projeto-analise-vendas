```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Visualizanado a minha Base de Dados Fictícia
df = pd.read_csv('../dados/vendas.csv')
df.head()
```

```python
#Trantando essa base de dados
df.info()
#Transformando data STR em data do tipo Date
df['data_compra'] = pd.to_datetime(df['data_compra'])
df.info()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(df["valor"])
plt.title("Distribuição de Valores de Compra")
plt.show()
```

```python
#Faturamento por categoria
df = df.groupby('categoria')['valor'].sum().reset_index()
faturamento = df.rename(columns={'valor': 'Valor Total'})
display(df)
```

```python
#Visualizando esse faturamento por categoria
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../dados/vendas.csv')
plt.figure(figsize=(8,5))
barplot = sns.barplot(x=df['categoria'],y=df['valor'], estimator= sum)
barplot.bar_label(barplot.containers[0],fontsize=10)
plt.title('Faturamento Por categoria')
plt.xlabel('Categoria')
plt.ylabel('Valor total')
plt.show()
```

```python
# Produtos Mais Vendidos

df = df[['produto']]

# 10 Produtos mais vendidos
qtd_produtos = df.value_counts().head(10)
# Top Três produtos mais vendidos
top_tres = qtd_produtos.sort_values(ascending=False).head(3)


plt.title('Top 3 produtos mais vendidos')
top_tres.plot(kind='bar',figsize=(10,5),color=['blue','green','purple'])
plt.title('Top 3 produtos Mais vendidos')
plt.xlabel('Produtos')
plt.ylabel('Numero de vendas')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Produto menos vendido e o mais vendido
menos_vendido = df['produto'].value_counts().idxmin()
quantidade_menos_vdo = df['produto'].value_counts().min()

mais_vendido = df['produto'].value_counts().idxmax()
quantidade_max_vdo = df['produto'].value_counts().max()


plt.figure(figsize=(5,4))
plt.bar(menos_vendido,quantidade_menos_vdo,color='gray' )
plt.title(f'Produto menos Vendido: {menos_vendido}')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.show()

plt.figure(figsize=(5,4))
plt.bar(mais_vendido,quantidade_max_vdo,color='green' )
plt.title(f'Produto menos Vendido: {mais_vendido}')
plt.xlabel('Produto')
plt.ylabel('Quantidade Vendida')
plt.show()
```

```python
# Tendencia Sazonal de compras
df['data_compra'] = pd.to_datetime(df['data_compra'])
df_mensal = df.set_index('data_compra').resample('ME').sum()
display(df_mensal)

plt.figure(figsize=(10,2))
plt.plot(df_mensal.index,df_mensal['valor'], marker='o', linestyle='-',color='red')
plt.title('Faturamento Por Mês')
plt.xlabel('Data')
plt.ylabel('Faturamento Total')
plt.grid()
plt.show()
```
